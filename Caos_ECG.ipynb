# **Entrop√≠a aproximada: Ejemplos particulares y ECG.**

```python
def ApEn(U, m, r) -> float:
    """Approximate_entropy."""

    def _maxdist(x_i, x_j):
        return max([abs(ua - va) for ua, va in zip(x_i, x_j)])

    def _phi(m):
        x = [[U[j] for j in range(i, i + m - 1 + 1)] for i in range(N - m + 1)]
        C = [
            len([1 for x_j in x if _maxdist(x_i, x_j) <= r]) / (N - m + 1.0)
            for x_i in x
        ]
        return (N - m + 1.0) ** (-1) * sum(np.log(C))

    N = len(U)

    return abs(_phi(m + 1) - _phi(m))
```

Se√±al regular --> Entrop√≠a menor

Se√±al aleatoria --> Entrop√≠a mayor

```python
m = 2
r = 0.1

# Se√±al Peri√≥dica.
N = 10**3
t = np.linspace(0, 300, N)
sp = np.sin(20*t)

print('N = ', len(sp))
print('ApEn_sp = ', ApEn(sp, m, r))

# Se√±al intermedia.
t1 = np.linspace(0,5,N)
si = np.sin(20*np.exp(-2*t1))

print('N = ', len(si))
print('ApEn_si = ', ApEn(si, m, r))

# Se√±al adicionandole un ruido.
t2 = np.linspace(0,300,N)
spa = np.sin(20*np.exp(t2))+random.randrange(0,5)

print('N = ', len(spa))
print('ApEn_spa = ',ApEn(spa, m, r))
```

Seg√∫n la teor√≠a, una se√±al regular deber√≠a tener una entrop√≠a menor respecto a la de una se√±al aleatoria. A modo formal, la se√±al peri√≥dica tiene m√°s patrones repetitivos que la se√±al aleatoria. En este escenario, podemos verificar dichos resultados, viendo que la entrop√≠a de la se√±al peri√≥dica con un valor $0.1387$ es menor que la de la se√±al con datos aleatorios $1.7488$.

# **Diagramas de Fase.**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Se√±al Peri√≥dica.
N = 10**3
t = np.linspace(0, 300, N)
sp = np.sin(20*t)

# t vs delta_x 
delta_sp = []
for n in range(1, N-1):
  sp[n] = sp[n+1] - sp[n]
  delta_sp.append(sp[n])

# x vs delta_x 
delta_sp = []
for n in range(1, N-1):
  sp[n] = sp[n+1] - sp[n]
  delta_sp.append(sp[n])

# x_n vs x_(n+1) 
sp_n1 = []
for n in range(1, N-1):
  sp_n1.append(sp[n+1])

fig, ((ax1, ax2) , (ax3, ax4)) = plt.subplots(2,2)
ax1.plot(t[:997],sp[:997])
ax1.set_title('Se√±al $SP= sen(20t)$')
ax1.set(xlabel='$t$', ylabel='SP')

ax2.plot(t[:997],delta_sp[:997])
ax2.set_title('$t$ vs $\Delta_{sp}$')
ax2.set(xlabel='$t$', ylabel='$\Delta_{sp}$')

ax3.plot(sp[:997],delta_sp[:997])
ax3.set_title('SP vs $\Delta_{sp}$')
ax3.set(xlabel='SP', ylabel='$\Delta_{sp}$')

ax4.plot(sp[:996],sp_n1[:996])
ax4.set_title('SP vs SP$_{n+1}$')
ax4.set(xlabel='SP', ylabel='SP$_{n+1}$')

fig.set_size_inches(18,10)
plt.show()
```
